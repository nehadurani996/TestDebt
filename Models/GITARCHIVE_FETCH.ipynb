```python
from itertools import zip_longest
import dask.dataframe as dd
import datetime
import itertools
import pandas as pd
import requests
import gzip
import json
import numpy as np
import re
```

```python
date_range = pd.date_range(start='01/01/2015', end='02/01/2023')
dates = pd.Series(date_range.date.astype("str"))
urls = []
for date in dates:
    for h in range(0, 24):
        urls.append(f"https://data.gharchive.org/{date}-{str(h)}.json.gz")

len(urls)
```

```python
urls = urls[:8]
```

```python
urls
```

```python
data_1 = dd.read_json(urls)
```

```python
data_all = data_1.compute()
```

```python
data_all
```

```python
data_issue = data_all[data_all["type"] == "IssuesEvent"]
```

```python
data_issue.reset_index(inplace = True)
```

```python
data_issue
```

```python
data_issue["payload"][0]
```

```python
records=[]
for i in range(len(data_issue)):

    id = data_issue["id"].values[i]
    type = data_issue["type"].values[i]
    created_at = data_issue["created_at"].values[i]
    repo = data_issue["repo"].values[i]["name"]
    repo_url = data_issue["repo"].values[i]["url"]
    action = data_issue["payload"].values[i]["action"]
    title = data_issue["payload"].values[i]["issue"]["title"]
    label = data_issue["payload"].values[i]["issue"]["labels"]
    if len(label) >= 1:
        labels = []
        for j in range(len(label)):
            labels.append(label[j]["name"])
        labels = " ".join(map(str, labels))
    else:
        labels = " ".join(map(str, label))
    body = data_issue["payload"].values[i]["issue"]["body"]
    dataframe_dict = {"id": id, "type": type, "created_at": created_at, "repo": repo, "repo_url": repo_url,
                      "action": action, "title": title, "labels": labels, "body": body}
    records.append(dataframe_dict)
```

```python
data_final_issue = pd.DataFrame.from_records(records)
data= data_final_issue.to_csv(f"issues_forTest_{str(created_at)}.csv", escapechar='\\')
```
