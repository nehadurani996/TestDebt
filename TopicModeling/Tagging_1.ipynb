```python

from langchain.llms import Ollama


from typing import Optional
from langchain.chains.openai_functions import create_structured_output_runnable
from langchain_community.chat_models import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.pydantic_v1 import BaseModel, Field

# Define the output schema
class IssueClassification(BaseModel):
    """Classification of GitHub issue text with category appended if related."""

    classification: str = Field(..., description="The classification of the GitHub issue text, with the category name appended if related, otherwise 'not_category'.")

# Define the prompt template
PROMPT_TEMPLATE = """\
Classify the following GitHub issue text as either '{category}' or 'not_{category}' based on its relevance to the category.

GitHub Issue Text:
{text}

Classification:
"""

# Initialize the ChatOpenAI
llm = ChatOpenAI(model="gpt-4-0125-preview", temperature=0)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are an advanced algorithm designed to analyse GitHub issue texts to identify the top 10 Common Vulnerabilities and Exposures (CVE) categories. For each category, provide a detailed description and a relevance score. Use the following top 10 CVEs as a guideline for your analysis:"),
        ("system", "1. CVE-2023-0001: Buffer Overflow - This vulnerability allows remote attackers to execute arbitrary code via a crafted input, score: 9.8."),
        ("system", "2. CVE-2023-0002: SQL Injection - A flaw that lets attackers run arbitrary SQL code on the database server, score: 9.5."),
        ("system", "3. CVE-2023-0003: Cross-Site Scripting (XSS) - Attackers can inject client-side scripts into web pages viewed by other users, score: 9.1."),
        ("system", "4. CVE-2023-0004: Insecure Deserialization - Allows attackers to execute arbitrary code by submitting crafted data, score: 8.8."),
        ("system", "5. CVE-2023-0005: XML External Entity (XXE) - External entities can be processed leading to unauthorized data disclosure, score: 8.5."),
        ("system", "6. CVE-2023-0006: Broken Authentication - Security measures are insufficient, allowing attackers to compromise user tokens or secrets, score: 8.2."),
        ("system", "7. CVE-2023-0007: Security Misconfiguration - Incorrect security settings make the application vulnerable, score: 8.0."),
        ("system", "8. CVE-2023-0008: Sensitive Data Exposure - Protecting personal data is inadequate, leading to risk of data breaches, score: 7.8."),
        ("system", "9. CVE-2023-0009: Insufficient Logging & Monitoring - Lack of or inadequate logging and monitoring allows breaches to go unnoticed, score: 7.5."),
        ("system", "10. CVE-2023-0010: Use After Free - Memory is improperly managed allowing attackers to execute arbitrary code, score: 7.2."),
        ("human", "Analyse the following GitHub issue text and provide the top 10 CVE categories that are most relevant, including a brief description and a relevance score for each."),
        ("human", "GitHub Issue Text: {text}"),
        ("human", "Your analysis should give me the  CVE ID  based on their relevance to the Github issue text"),
    ]
)

# Example usage

chain = create_structured_output_runnable(IssueClassification, llm, prompt)
```

```python
#### INDEXING ####

# Load Documents
loader = JSONLoader(
    file_path="extracted_data.json",
    jq_schema=".",
    text_content=False

)
docs = loader.load()
```

```python
docs
```

```python

# Split
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
splits = text_splitter.split_documents(docs)

# Embed
vectorstore = Chroma.from_documents(documents=splits,
                                    embedding=OpenAIEmbeddings())

retriever = vectorstore.as_retriever()

#### RETRIEVAL and GENERATION ####
```

```python



# Prompt
prompt = hub.pull("rlm/rag-prompt")




# LLM
llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0)

# Post-processing
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

# Chain
rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

# Question
rag_chain.invoke("here is the CVE ID   CVE-2024-0007")
```
