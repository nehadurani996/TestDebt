```python
import pandas as pd
# Load the CSV data into a DataFrame
df1 = pd.read_csv("TechNotTechRes.csv")
df1 = df1.rename(columns={'predicted_label_numeric': 'predicted_label_Technical'})
df1.head()
```

```python
import pandas as pd
# Load the CSV data into a DataFrame
df2 = pd.read_csv("TestNotTestRes_bert.csv")
df2 = df2.rename(columns={'clean_text': 'text_clean','actual_type':'label','predicted_type_numeric':'predicted_label_Test'})
df2.head()
```

```python
merged_df = pd.merge(df1[['text_clean', 'label', 'predicted_label_Technical']],
                     df2[['text_clean', 'label', 'predicted_label_Test']],
                     on=['text_clean', 'label'])

# Display the first few rows of the merged DataFrame to verify
merged_df.head()
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import time
start_time = time.time()
# Create a new column 'final_predicted' that checks if both predictions are the same
# If they are the same, it takes that value; otherwise, it assigns 0
merged_df['final_predicted'] = merged_df.apply(
    lambda row: row['predicted_label_Technical'] if row['predicted_label_Technical'] == row['predicted_label_Test'] else 0, axis=1)

# Now, compare 'final_predicted' with the actual label ('label') to find accuracy and F1 score
actual_labels = merged_df['label']
predicted_labels = merged_df['final_predicted']
merged_df.head(100)
```

```python
# Calculate accuracy
accuracy = accuracy_score(actual_labels, predicted_labels)
f1 = f1_score(actual_labels, predicted_labels, average='weighted')  # Use 'binary' for binary classification
precision = precision_score(actual_labels, predicted_labels, average='weighted')  # Use 'binary' for binary classification
recall = recall_score(actual_labels, predicted_labels, average='weighted')  # Use 'binary' for binary classification

# End measuring runtime
end_time = time.time()
runtime = end_time - start_time

# Display the metrics and runtime
print(f"Accuracy: {accuracy}")
print(f"F1 Score: {f1}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"Runtime: {runtime} seconds")
```
