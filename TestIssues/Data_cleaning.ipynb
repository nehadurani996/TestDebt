```python
#Give test a label by number.
# 'Label encoding'. Makes is easier for machine learning models to work with categorical data.
test["label"] = 0
test["class"] = "test"
test.head()
```

```python
not_test["label"] = 1
not_test["class"] = "not_test" 
not_test.head()
```

```python
# Drop duplicates by the content of the body
test = test.drop_duplicates(subset=['title'], keep='last')
test.dropna(inplace=True)
test.reset_index(inplace=True)
test.drop(columns=["index"] , inplace= True)
test["class"].value_counts()
```

```python
# Drop duplicates by the content of the body
not_test = not_test.drop_duplicates(subset=['title'], keep='last')
not_test.dropna(inplace=True)
not_test.reset_index(inplace=True)
not_test.drop(columns=["index"] , inplace= True)
not_test["class"].value_counts()
```

```python
test.info()
```

```python
label_counts = test["class"].value_counts()
label_counts_nt = not_test["class"].value_counts()
print(label_counts)
not_test_count = label_counts_nt["not_test"]
print(not_test_count)
t_count = label_counts["test"]
t_count
```

```python
not_test = not_test.sample(frac=t_count/not_test_count, random_state=42, replace= True)
not_test
```

```python
all_debt = pd.concat([test,not_test] , ignore_index = True)
all_debt.tail()
```

```python
# Does this need to be more balanced?
all_debt["label"].value_counts()
```

```python
print(all_debt["title"][0])
print(all_debt["body"][0])
```

```python
# Copy content of body to a new col named text
all_debt["text"] = all_debt["title"] + all_debt["body"]
all_debt.tail()
```

```python
all_debt["text"][0]
```

```python
# Make a new dataframe with only text, label and class cols.
all_debt_subset = all_debt[["text" , "label" , "class"]]
all_debt_subset
```

```python
# Convert to string
all_debt_subset["text_str"] = all_debt_subset['text'].astype(str)
```

```python
all_debt_subset
```

```python
all_debt_subset["text_clean"] = all_debt_subset["text_str"].map(preprocess_text)
```

```python
# Make a subset with text_clean and label
test_label_text= all_debt_subset[["text_clean" , "label"]]
test_label_text
```

```python
# Assuming all_debt_subset is your original DataFrame
# Filter rows where 'class' is 'test' and select only 'text_clean' and 'class' columns
test_label_text_tp = all_debt_subset[all_debt_subset['class'] == 'test'][['text_clean', 'class']]

# Display the first few rows of the filtered DataFrame to verify
test_label_text_tp.head()
```

```python
# Need to dropna here since cleaning function returns NaN for not english text.
test_label_text_tp.dropna(inplace=True)
test_label_text_tp.reset_index(inplace=True)
test_label_text.dropna(inplace=True)
test_label_text.reset_index(inplace=True)
test_label_text.drop(columns=["index"] , inplace= True)

test_label_text
```

```python
test_label_text_tp.to_csv('tagging.csv', index=False)
test_label_text_tp.head()
```

```python
# 10% of the data for testing
test_df = test_label_text.sample(frac=0.1, random_state=42)
# Remove test data from test_text
test_label_text = test_label_text.drop(test_df.index)
test_df
```

```python
test_df = test_df.reset_index(drop=True)
test_label_text = test_label_text.reset_index(drop=True) 
test_label_text
```
